```python
# _k, _n = 2, 5
# _c = [6, 7]

# _k, _n = 1, 5
# _c = [9]

_k, _n = 5, 2
_c = [11, 10, 7, 5, 18]

```

```python
k, n = map(int, input().split())
c = list(map(int, input().split()))

k, n = _k, _n
c = _c

dp = [0] * n
dp[0] = c[0]

for i in range(1, n):
    dp[i] = -float('inf')
    for j in range(max(0, i-k), i):
        dp[i] = max(dp[i], dp[j] + c[i % k])

print(max(dp))
```

```python
k, n = map(int, input().split())
c = list(map(int, input().split()))

c.sort(reverse=True)
result = 0
for shift in range(k):
    _temp = 0
    for index in range(n):
        _temp += (c[(index + shift) % k])
    if _temp >= result:
        result = _temp
print(result)
```
